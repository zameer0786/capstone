```python
from rake_nltk import Rake
r = Rake()
```

```python
key=[]
for i in Article['News']:
    #Example
    a=r.extract_keywords_from_text(i)
    b=r.get_ranked_phrases()
    c=r.get_ranked_phrases_with_scores()
    key.append(c)
    
```

# we calculate the means scores for each article and take the words who have their scores > mean*1.5

```python
score=[]
avg=[]
for i in key:
    score=[]
    for j in i:
        #print(j[0])
        score.append(j[0])
    avg.append(np.mean(score)*1.5)
```

```python
Article['Avg_score'] = avg
```

```python
main=[]
for i,j in zip(key,Article['Avg_score']):
    filt=[]
    
   
    
    
    for q in range(0,len(i)):
        if(i[q][0]>j):
                
            filt.append((i[q][1]))
    main.append(str(filt))
            
    
   
```
