# REPLACE YEARS BASED ON INCOME

# 10000 to 50000---> 10 
# 50000 to 100000---> 15
# 100000 to 200000---> 20
# 200000 to 300000---> 25
# 300000 to 400000---> 30

```python
mask = (df['emp_length'] == '10+') & ((df['annual_inc'] >= 10000) & (df['annual_inc'] < 500000))
df['emp_length'][mask] = 10
```

```python
mask = (df['emp_length'] == '10+') & ((df['annual_inc'] >= 500000) & (df['annual_inc'] < 1000000))
df['emp_length'][mask] = 15
```

```python
mask = (df['emp_length'] == '10+') & ((df['annual_inc'] >= 1000000) & (df['annual_inc'] < 2000000))
df['emp_length'][mask] = 20
```

```python
mask = (df['emp_length'] == '10+') & ((df['annual_inc'] >= 2000000) & (df['annual_inc'] < 3000000))
df['emp_length'][mask] = 25
```

```python
mask = (df['emp_length'] == '10+') & ((df['annual_inc'] >= 3000000) & (df['annual_inc'] < 4000000))
df['emp_length'][mask] = 30
```
